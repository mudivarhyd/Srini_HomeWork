```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_df, school_df, how="left", on="school_name")
school_data_complete_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = school_df["School ID"].count()
# Calculate the total number of students
total_students = student_df["Student ID"].count()
# Calculate the total budget
total_budget = school_df["budget"].sum()
# Calculate the average math score
average_math_score = student_df["math_score"].mean()
# Calculate the average reading score
average_reading_score = student_df["reading_score"].mean()
# Calculate overall passing rate
overall_passing_rate = (average_math_score + average_reading_score) / 2
# Calculate the percentage of students with passing math score >= 70
math_score_greater = student_df.loc[student_df["math_score"] >= 70]["Student ID"].count()/total_students*100
# Calculate the percentage of students with passing reading score >= 70
reading_score_greater = student_df.loc[student_df["reading_score"] >= 70]["Student ID"].count()/total_students*100
# Create a District summary data frame
dist_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                                "Total Students": "{0:,.0f}".format(total_students),
                                "Total Budget": "${0:,.2f}".format(total_budget),
                                "Average Math Score": "{0:,.6f}".format(average_math_score),
                                "Average Reading Score": "{0:,.6f}".format(average_reading_score),
                                "%Passing Math": "{0:,.6f}".format(math_score_greater),
                                "%Passing Reading": "{0:,.6f}".format(reading_score_greater),
                                "%Overall Passing Rate": "{0:,.6f}".format(overall_passing_rate)})
dist_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# School Summary
# Merged data is grouped by school name and school type
school_type_group = school_data_complete_df.groupby(["school_name"])
# School Types by school
type_group = school_type_group["type"].min()
# Count of total schools by shool name and school type
#total_students_group = school_type_group["School ID"].count().sort_values(ascending = False)
total_students_group = school_type_group["Student ID"].count()
# Total school budget by school name and school type
total_budget_group = school_type_group["budget"].mean()
# Per student budget by school name and school type
total_per_student = total_budget_group/total_students_group
# Average math scores by school name and school type
avg_math_score = school_type_group["math_score"].mean()
# Average reading scores by school name and school type
avg_reading_score = school_type_group["reading_score"].mean()
# Percent pass math scores by school name and school type
percent_math_score = school_data_complete_df[school_data_complete_df["math_score"] >= 70].groupby(["school_name"])\
                    ["Student ID"].count() / total_students_group * 100
#print(percent_math_score)
# Percent pass reading scores by school name and school type
percent_reading_score = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]\
                    .groupby(["school_name"])["Student ID"].count() / total_students_group*100
# Overall passing rate
passing_rate = (percent_math_score + percent_reading_score)/2

# Create a dataframe for the school summary
school_summary_df = pd.DataFrame({"School Type": type_group,
                                  "Total Students": total_students_group.map('{0:,.0f}'.format),
                                  "Total School Budget": total_budget_group.map('${:,.2f}'.format),
                                  "Per Student Budget": total_per_student.map('${:,.2f}'.format),
                                  "Average Math Score": avg_math_score.map('{0:,.6f}'.format),
                                  "Average Reading Score": avg_reading_score.map('{0:,.6f}'.format),
                                  "% Passing Math": percent_math_score.map('{0:,.6f}'.format),
                                  "% Passing Reading": percent_reading_score.map('{0:,.6f}'.format),
                                  "% Overall Passing Rate": passing_rate.map('{0:,.6f}'.format)})
school_summary_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Scores by school spending

school_data_complete_df["spending_bin"] = pd.cut(school_data_complete_df["budget"] / \
                                            school_data_complete_df["size"], spending_bins, labels = group_names)
                                                 
#school_data_complete_df.head(2)

# Group by spending bins
school_spending_df = school_data_complete_df.groupby("spending_bin")
# Average math score by spending bins
avg_math_score = school_spending_df["math_score"].mean()
# Average reading score by spending bins
avg_reading_score = school_spending_df["reading_score"].mean()
# Percent pass math scores by spending bins
percent_math_score = school_data_complete_df[school_data_complete_df["math_score"] >= 70].groupby(["spending_bin"])\
                    ["Student ID"].count() / school_spending_df["Student ID"].count() * 100
# Percent pass reading scores by school name and school type
percent_reading_score = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]\
                    .groupby(["spending_bin"])["Student ID"].count() / school_spending_df["Student ID"].count() * 100
# Overall passing rate
passing_rate = (percent_math_score + percent_reading_score)/2

# Create a dataframe for the spending summary by bin
spending_summary_df = pd.DataFrame({"Average Math Score": avg_math_score.map('{0:,.6f}'.format),
                                  "Average Reading Score": avg_reading_score.map('{0:,.6f}'.format),
                                  "% Passing Math": percent_math_score.map('{0:,.6f}'.format),
                                  "% Passing Reading": percent_reading_score.map('{0:,.6f}'.format),
                                  "% Overall Passing Rate": passing_rate.map('{0:,.6f}'.format)})
spending_summary_df.index.name = "Spending Ranges (Per Student)"                                  
spending_summary_df
```
